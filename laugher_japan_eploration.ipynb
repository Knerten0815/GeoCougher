Ich erstelle eine Spalte uuid die sich aus dem file_name der Audio datei ergibt. Sie beschreibt um welchen Sprecher es sich handelt (erster Tiel der ID) und welche Rolle er spricht. Im Datensatz sprechen die selben Sprecher teilweise mehrere Rollen und zur Unterschiedung dieser dient die Ziffier.
Es sei erwähnt, dass ein und der selbe Sprecher in der selben Rolle verschiedene Lacharten inszeniert, welshalb es eine uuid mehrfach geben kann.

```python
def extract_id(file_path):
    parts = file_path.split('/')
    speaker_id = parts[1]
    role_id = parts[2]
    return f"{speaker_id}-{role_id}"

data_japan['uuid'] = data_japan['file_name'].apply(extract_id)

data_japan.head()
```

```python
data_japan = data_japan.reindex(columns=['uuid','file_name', 'transcription'])
data_japan = data_japan.drop(columns=['transcription'])
data_japan['language'] = 'JA'
data_japan['iso639_3'] = 'jpn'
data_japan['iso639_5'] = 'jpx'
data_japan['family1'] = 'jpx'
data_japan['country'] = 'Japan'

data_japan.head()
```

Aufgrund der Struktur der Dateibenennung: {speaker_ID}_{class}_{trial}_{sex}_{age}_{location}_{quality}_{noise}.wav extrahiere ich alle IDs, welche tatsächlich unter den lauger Audiofiles vertreten sind.

```python
audio_dir = "Datasets/deeplyinc_Nonverbal-Vocalization-Dataset_laughing"
audio_files = os.listdir(audio_dir)
audio_ids = set([filename.split('_')[0] for filename in audio_files])
data_south_korea = data_south_korea[data_south_korea.index.isin(audio_ids)]
```
